```python
from twon_lss.utility.eval import RunEvaluation
from twon_lss.utility.llm import LLM
import dotenv
import joblib
```

```python
ENV = dotenv.dotenv_values("../" * 3 + ".env")
```

```python
embedding_model = LLM(api_key=ENV["HF_TOKEN"], model="mxbai-embed-large-v1", url="https://router.huggingface.co/hf-inference/models/mixedbread-ai/mxbai-embed-large-v1/pipeline/feature-extraction")
```

```python
base_barabasi_semantic = RunEvaluation(path = "runs/Qwen3-8B-1028-barabasi_albert-Seed42-SemanticSimilarityRanker-20251117_125126/feed.json", name="base_barabasi_semantic", embedding_model=embedding_model)
base_barabasi_random = RunEvaluation(path = "runs/Qwen3-8B-1028-barabasi_albert-Seed42-RandomRanker-20251117_083443/feed.json", name="base_barabasi_random", embedding_model=embedding_model)
base_complete_random = RunEvaluation(path = "runs/Qwen3-8B-1028-complete-Seed42-RandomRanker-20251117_082741/feed.json", name="base_complete_random", embedding_model=embedding_model)
base_complete_semantic = RunEvaluation(path = "runs/Qwen3-8B-1028-complete-Seed42-SemanticSimilarityRanker-20251117_082729/feed.json", name="base_complete_semantic", embedding_model=embedding_model)

agent_complete_random = RunEvaluation(path = "runs/TWON-Agent-8B-1028-complete-Seed42-RandomRanker-20251114_123558/feed.json", name="agent_complete_random", embedding_model=embedding_model)
agent_barabasi_random = RunEvaluation(path = "runs/TWON-Agent-8B-1028-barabasi_albert-Seed42-RandomRanker-20251114_122936/feed.json", name="agent_barabasi_random", embedding_model=embedding_model)
agent_complete_semantic = RunEvaluation(path = "runs/TWON-Agent-8B-1028-complete-Seed42-SemanticSimilarityRanker-20251115_161142/feed.json", name="agent_complete_semantic", embedding_model=embedding_model)
agent_barabasi_semantic = RunEvaluation(path = "runs/TWON-Agent-8B-1028-barabasi_albert-Seed42-SemanticSimilarityRanker-20251115_161610/feed.json", name="agent_barabasi_semantic", embedding_model=embedding_model)

snapshot = RunEvaluation(path = "../data/snapshot.json", name="snapshot", embedding_model=embedding_model)

runs = [
    snapshot,
    base_complete_random,
    base_barabasi_random,
    base_complete_semantic,
    base_barabasi_semantic,
    agent_complete_random,
    agent_barabasi_random,
    agent_complete_semantic,
    agent_barabasi_semantic
]
```

```python
jigsaw_classifier = joblib.load("../classifier/jigsaw.joblib")
goemotions_classifier = joblib.load("../classifier/goemotions.joblib")
davidson_classifier = joblib.load("../classifier/davidson.joblib")
semeval_a_classifier = joblib.load("../classifier/semeval_task_a.joblib")
semeval_b_classifier = joblib.load("../classifier/semeval_task_b.joblib")
semeval_c_classifier = joblib.load("../classifier/semeval_task_c.joblib")

sentiment_classifier = joblib.load("../classifier/tweetsentiment_sentiment.joblib")
intimacy_classifier = joblib.load("../classifier/tweetintimacy_intimacy.joblib")
topic_classifier = joblib.load("../classifier/tweettopic_topic.joblib")

classifiers = [
    goemotions_classifier,
    jigsaw_classifier,
    davidson_classifier,
    semeval_a_classifier,
    semeval_b_classifier,
    semeval_c_classifier,
    sentiment_classifier,
    intimacy_classifier,
    topic_classifier
]
```

```python
RunEvaluation.describe_runs(runs = runs)
```

```python
RunEvaluation.compare_runs_content(runs=runs, classifiers=classifiers)
```

```python
RunEvaluation.compare_runs_topics(runs=runs, n_topics=15)
```

```python
RunEvaluation.compare_runs_content_over_time(runs=runs, classifiers=[sentiment_classifier["classifier"]], running_average=True, window_size=15)
```
