```python
from twon_lss.utility.eval import RunEvaluation
from twon_lss.utility.llm import LLM
import dotenv
import joblib
```

```python
ENV = dotenv.dotenv_values("../" * 3 + ".env")
```

```python
embedding_model = LLM(api_key=ENV["HF_TOKEN"], model="mxbai-embed-large-v1", url="https://router.huggingface.co/hf-inference/models/mixedbread-ai/mxbai-embed-large-v1/pipeline/feature-extraction")
```

```python
complete_random = RunEvaluation(path = "runs/TWON-Agent-8B-1028-complete-Seed42-RandomRanker-20251114_123558/feed.json", name="complete_random", embedding_model=embedding_model)
barabasi_random = RunEvaluation(path = "runs/TWON-Agent-8B-1028-barabasi_albert-Seed42-RandomRanker-20251114_122936/feed.json", name="barabasi_random", embedding_model=embedding_model)
complete_semantic = RunEvaluation(path = "runs/TWON-Agent-8B-1028-complete-Seed42-SemanticSimilarityRanker-20251115_161142/feed.json", name="complete_semantic", embedding_model=embedding_model)
barabasi_semantic = RunEvaluation(path = "runs/TWON-Agent-8B-1028-barabasi_albert-Seed42-SemanticSimilarityRanker-20251115_161610/feed.json", name="barabasi_semantic", embedding_model=embedding_model)

snapshot = RunEvaluation(path = "../data/snapshot.json", name="snapshot", embedding_model=embedding_model)
```

```python
RunEvaluation.describe_runs([complete_random, barabasi_random, complete_semantic, barabasi_semantic, snapshot])
```

```python
jigsaw_classifier = joblib.load("../classifier/jigsaw.joblib")
goemotions_classifier = joblib.load("../classifier/goemotions.joblib")
davidson_classifier = joblib.load("../classifier/davidson.joblib")
semeval_a_classifier = joblib.load("../classifier/semeval_task_a.joblib")
semeval_b_classifier = joblib.load("../classifier/semeval_task_b.joblib")
semeval_c_classifier = joblib.load("../classifier/semeval_task_c.joblib")
```

```python
fig = RunEvaluation.compare_runs_content(runs=[snapshot, complete_random, barabasi_random, complete_semantic, barabasi_semantic], classifiers=[goemotions_classifier, jigsaw_classifier, davidson_classifier, semeval_a_classifier, semeval_b_classifier, semeval_c_classifier])
```

```python
RunEvaluation.compare_runs_topics(runs=[snapshot, complete_random, barabasi_random, complete_semantic, barabasi_semantic], n_topics=30)
```

```python
RunEvaluation.compare_runs_over_time([snapshot, complete_random, barabasi_random, complete_semantic, barabasi_semantic], [davidson_classifier["classifier"]], running_average=True, window_size=15)
```
