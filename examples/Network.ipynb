```python
import typing

import networkx
import matplotlib.pyplot as plt
```

```python
from twon_lss.schemas import Network, User
```

```python
N: int = 50
```

```python
users: typing.List[User] = [User() for _ in range(N)]
network = Network.from_graph(networkx.random_regular_graph(5, N), users)
print(network.root)
```

```python
plt.figure(figsize=(12, 12), dpi=300)

networkx.draw_networkx(network.root, with_labels=True, width=0.5, font_size=6)
```
