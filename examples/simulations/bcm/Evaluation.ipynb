```python
import json

import pandas
import seaborn
```

```python
seaborn.set_style()
```

```python
data = pandas.concat(
    [
        pandas.Series(data["memory"], name=idx)
        for idx, data in json.load(open("output/individuals.json")).items()
    ],
    axis=1,
)
data
```

```python
(
    seaborn.lineplot(
        df := (
            data.reset_index().melt(id_vars=["index"]).rename(columns={"index": "step"})
        ),
        **(axis := dict(x="step", y="value")),
        hue="variable",
        lw=0.5,
        legend=False,
    ),
    # seaborn.lineplot(
    #     df,
    #     **axis,
    #     lw=2.0,
    #     legend=False,
    #     err_kws={'linestyle': '--', 'hatch': 'xxx', 'alpha': 0.4}
    # )
)
```
